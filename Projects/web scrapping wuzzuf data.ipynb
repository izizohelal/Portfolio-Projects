```python
#importing libraries
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

```python
#inputting the url
url = 'https://wuzzuf.net/search/jobs/?a=navbl&q=data%20analyst'
```

```python
#create a client based request to get the url
client = urlopen(url)
```

```python
#getting the html code of the full page
html = client.read()
html
```

```python
#closing the request
client.close()
```

```python
#creating html parser using beautiful soup 
soup = bs(html,'html.parser')
```

```python
#creating a container for needed data 
containers = soup.find_all('div',{'class':'css-1gatmva e1v1l3u10'})
len(containers)
```

```python
bs.prettify(containers[0])
```

```python
f = open('wuzuff-dataanalyst.csv','w')
header = 'job_title,company_name,job_type\n'
f.write(header)
```

```python
for container in containers:
    jtitle = container.findAll('h2',{'class':'css-m604qf'})
    job_title = jtitle[0].text.strip()
    
    cname = container.findAll('a',{'class':'css-17s97q8'})
    company_name = cname[0].text.strip()
    
    jtype = container.findAll('span',{'class':'css-1ve4b75'})
    job_type = jtype[0].text.strip()
    
    #print(job_title)
    #print(company_name)
    #print(job_type)
    #print()
    f.write(job_title + ',' + company_name + ',' + job_type + '\n')
f.close()    
```
